### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries=list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
#max_infection_rates# it will so the list

corona_dataset_aggregated["max_infection_rate"]=max_infection_rates
```

```python
deaths_reported=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
recovered_cases=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
# now extracting all the columns using the .keys() function 

cols=confirmed_cases.keys()
cols
```

```python
# extracting only the dates from confirmed_cases
confirmed=confirmed_cases.loc[:,cols[4]:cols[-1]]
```

```python
# extracting only the dates from deaths_reported
deaths=deaths_reported.loc[:,cols[4]:cols[-1]]
```

```python
# extracting only the dates from recoverd_cases
recovered=recovered_cases.loc[:,cols[4]:cols[-1]]
```

```python
confirmed.head()
```

```python
deaths.head()
```

```python
recovered.head()
```

```python
# finfing the total confirmed cases,death cases and recoverd cases and append them to an 4 empty lists
# Also,calculating the total mortality rate which is the death_sum/confirmed cases

dates=confirmed.keys()
world_cases=[]
total_deaths=[]
mortality_rate=[]
total_recovered=[]

for i in dates:
    confirmed_sum=confirmed[i].sum()
    death_sum=deaths[i].sum()
    recoverd_sum=recovered[i].sum()
    world_cases.append(confirmed.sum)
    total_deaths.append(death_sum)
    mortality_rate.append(death_sum/confirmed_sum)
    total_recovered.append(recoverd_sum)
    
    
```
